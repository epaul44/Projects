```python
import pandas as pd
import numpy as np
import os
import ast
import matplotlib.pyplot as plt
import cv2
```

```python
df_epi = pd.read_csv("EpiduralHemorrhageRemovedFlags.csv")

cond1 = df_epi["Correct Label"].notna()
cond2 = df_epi["Correct Label"].isna()

v1 = df_epi["Correct Label"]
v2 = df_epi["Majority Label"]

df_epi["OutLabel"] = np.select([cond1, cond2], [v1, v2])
df_epi
```

```python
df_subdur = pd.read_csv("SubduralHemorrhageRemovedFlags.csv")

cond1 = df_subdur["Correct Label"].notna()
cond2 = df_subdur["Correct Label"].isna()

v1 = df_subdur["Correct Label"]
v2 = df_subdur["Majority Label"]

df_subdur["OutLabel"] = np.select([cond1, cond2], [v1, v2])
df_subdur
```

```python
df_subarc = pd.read_csv("SubarachnoidHemorrhageRemovedFlags.csv")

cond1 = df_subarc["Correct Label"].notna()
cond2 = df_subarc["Correct Label"].isna()

v1 = df_subarc["Correct Label"]
v2 = df_subarc["Majority Label"]

df_subarc["OutLabel"] = np.select([cond1, cond2], [v1, v2])
df_subarc
```

```python
df_intra = pd.read_csv("IntraparenchymalHemorrhageRemovedFlags.csv")

cond1 = df_intra["Correct Label"].notna()
cond2 = df_intra["Correct Label"].isna()

v1 = df_intra["Correct Label"]
v2 = df_intra["Majority Label"]

df_intra["OutLabel"] = np.select([cond1, cond2], [v1, v2])
df_intra
```

```python
df_multi = pd.read_csv("MultipleHemorrhageRemovedFlags.csv")

cond1 = df_multi["Correct Label"].notna()
cond2 = df_multi["Correct Label"].isna()

v1 = df_multi["Correct Label"]
v2 = df_multi["Majority Label"]

df_multi["OutLabel"] = np.select([cond1, cond2], [v1, v2])
df_multi
```

```python
df_masks = pd.concat([df_multi, df_intra, df_subarc, df_subdur, df_epi], ignore_index=True)
df_masks
```

```python
df_masks = df_masks[["Case ID", "Origin", "All Labels", "OutLabel"]]
df_masks = df_masks[df_masks["OutLabel"] != '[]']
df_masks = df_masks[df_masks["OutLabel"] != '[[], []]']
df_masks["OutLabel"] = df_masks["OutLabel"].apply(lambda x: ast.literal_eval(x))
df_masks["All Labels"] = df_masks["All Labels"].apply(lambda x: ast.literal_eval(x))
```

```python
def flat(x):
    res = []
    for i in x:
        if type(i) is list:
            res.extend(i)
    return res if len(res) > 1 else x

def pull(x):
    res = []
    for i in x:
        if type(i) is dict:
            res.append((i['x'], i['y']))
    return res if len(res) > 1 else x
```

```python
df_masks["OutLabel"] = df_masks["OutLabel"].apply(lambda x: flat(x))
df_masks["All Labels"] = df_masks["All Labels"].apply(lambda x: flat(x))
df_masks
```

```python
df_masks["OutLabel"] = df_masks["OutLabel"].apply(lambda x: pull(x))
df_masks["All Labels"] = df_masks["All Labels"].apply(lambda x: pull(x))
df_masks
```

```python
df = df_masks.groupby('Case ID', sort=False).agg({'Origin': lambda x: x, 'All Labels': lambda x: x,'OutLabel': lambda x: x})
df = df.reset_index()
df
```

```python
cnt = 0
for lopts in df["OutLabel"]:
    orgjpeg = df['Origin'][cnt][0] if type(df['Origin'][cnt]) is np.ndarray else df['Origin'][cnt]
    fname = orgjpeg.split('.')[0] + 'mask' + '.png'
    canvas = np.zeros((512, 512,3), np.uint8)
    ouch = cnt
    if type(lopts) is np.ndarray:
        for p in lopts:
            if p != [[]]:
                pts = np.array([[int(x*512), int(y*512)] for (x,y) in p])
                cv2.drawContours(canvas, [pts],-1, (0,255,0), 3)
                cv2.fillPoly(canvas, [pts], color=(255,255,255))
    elif lopts != [[]]:
            pts = np.array([[int(x*512), int(y*512)] for (x,y) in lopts])
            cv2.drawContours(canvas, [pts],-1, (0,255,0), 3)
            cv2.fillPoly(canvas, [pts], color=(255,255,255))
    os.chdir(directory)
    cv2.imwrite(rf'{fname}', canvas)
    cnt += 1
```
